```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

# Copying an image from one directory to another

```python
present_image_path = r'/grayscale.jpg'
```

```python
destination_directory = r'/target'
```

```python
#reading the image from current path
img = cv2.imread('grayscale.jpg')
```

```python
#change directory to destination directory/folder 
os.chdir('/content/target/')
```

```python
#listing items from root directory
print(os.listdir('/')) 
```

```python
#listing items from destination directory
print(os.listdir('/content/target'))
```

```python
#writing (moving) the image to 
new_image_name = 'new_grayscale.jpg'
cv2.imwrite(new_image_name, img)
```

```python
print(os.listdir('/content/target'))
```

```python
#displaying the new image
new_image = cv2.imread('new_grayscale.jpg')
plt.imshow(new_image)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
original_grayscale_image = cv2.imread('grayscale.jpg')

print(original_grayscale_image.shape)
```

```python
# Extracting the height and width of an image
height, width = original_grayscale_image.shape[:2]
# Displaying the height and width
print("Height = {},  Width = {}".format(height, width))
```

# Drawing a rectangle on any image

```python
'''Syntax: cv2.rectangle(image, start_point, end_point, color, thickness)

start_point/end_point: These coordinates are represented as tuples of two values i.e. (X coordinate value, Y coordinate value).
color: It is the color of border line of rectangle to be drawn. For BGR, we pass a tuple. eg: (255, 0, 0) for blue color.
thickness: It is the thickness of the rectangle border line in px. Thickness of -1 px will fill the rectangle shape by the specified color.'''

start_point = (0,0)
end_point = (300,300)
color = (255, 0, 0)
thickness = -1
image_with_rectangle = cv2.rectangle(original_grayscale_image, start_point, end_point, color, thickness)
```

```python
plt.imshow(image_with_rectangle)
plt.show()
```

```python
image_with_rectangle = cv2.rectangle(image_with_rectangle, (0,300), (300,height), (0,255,0), -1)
plt.imshow(image_with_rectangle)
plt.show()
```

```python
image_with_rectangle = cv2.rectangle(image_with_rectangle, (700,0), (width,300), (0,0,255), -1)
plt.imshow(image_with_rectangle)
plt.show()
```

```python
image_with_rectangle = cv2.rectangle(image_with_rectangle, (700,300), (width,height), (255,255,0), -1)
plt.imshow(image_with_rectangle)
plt.show()
```

# Drawing circle on an image

```python
''' Syntax: cv2.circle(image, center_coordinates, radius, color, thickness)

center_coordinates: represented as tuples of two values i.e. (X coordinate value, Y coordinate value). 
radius: It is the radius of the circle. 
color: It is the color of the borderline of a circle to be drawn. For BGR, we pass a tuple. eg: (255, 0, 0) for blue color. 
thickness: It is the thickness of the circle border line in px. Thickness of -1 px will fill the circle shape by the specified color.
'''
```

```python
original_grayscale_image = cv2.imread('grayscale.jpg')
```

```python
image_with_other_shapes= cv2.circle(original_grayscale_image, (100,300), 50, (255,0,0), -1)
plt.imshow(image_with_other_shapes)
plt.show()
```

#Drawing a straight line

```python
''' Syntax: cv2.line(image, start_point, end_point, color, thickness) '''

image_with_other_shapes = cv2.line(image_with_other_shapes, (0,400), (200,400), (0,255,0), 10)
plt.imshow(image_with_other_shapes)
plt.show()
```

```python
#Drawing a diagonal through the image - bottom left corner to top right corner
start_point = (0,height)
end_point = (width,0)
image_with_other_shapes = cv2.line(image_with_other_shapes, start_point, end_point, (255,255,0), 10)
plt.imshow(image_with_other_shapes)
plt.show()
```
